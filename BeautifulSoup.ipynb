```python
#Descargando una web
import requests

url = 'http://example.webscraping.com'
examplews = requests.get(url)
examplews.status_code
```

```python
print(examplews.text)
```

```python
#Usando BeautifulSoup
#pip install beautifulsoup4
from bs4 import BeautifulSoup
```

```python
exampbs = BeautifulSoup(examplews.text, 'lxml')
```

```python
print(exampbs.prettify())
```

```python
exampbs.find('a', attrs={'href':'/places/default/view/Afghanistan-1'})
```

```python
tds = exampbs.find_all('td')
tds
```

```python
for td in tds:
    print(td.find('a').text)
```
